```r
library(ggplot2)
```

```r
ggplot(Data)+geom_histogram(aes(x=Usage),bins=20,col='purple',fill='orange',linetype="dashed")
```

#### There are more states with less power usage. 

```r
options(repr.plot.width=27 ,repr.plot.height= 10)
aggData <- aggregate(x = Data$Usage,
                     by=list(States = Data$States),
                     FUN = mean)
ggplot(data = aggData, aes(x = States, y = prop.table(stat(aggData$x)), fill = States, label = scales::percent(prop.table(stat(aggData$x))))) +
  geom_bar(stat="identity", position = "dodge") + 
  geom_text(stat = 'identity', position = position_dodge(.9),  vjust = -0.5, size = 3) + 
  scale_y_continuous(labels = scales::percent) + 
  labs(x = 'States', y = 'Usage in Percentage', fill = 'States') + 
  ggtitle("Mean contribution to Usage by State")
```

#### On an average Maharashtra(12.1%) followed by Gujrat(9.5%) consumes more power.
